```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y, 'r*-')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'r-*')
plt.subplot(1,3,2)
plt.plot(y,x,'g-*')
plt.subplot(1,3,3)
plt.plot(y**2,x,'b-*')
```

## Object Oriented method

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,0.9,0.9])
axes2 = fig.add_axes([1.0,0,0.9,0.9])
axes3 = fig.add_axes([2.0,0,0.9,0.9])

axes1.plot(x,y,'b')
axes2.plot(y,x,'r')
axes3.plot(x**2,y,'g')
```

### Subplot()

```python
fig,axes = plt.subplots(1,2)

axes[0].plot(x,y,'r')
axes[0].set_title('First Plot')

axes[1].plot(y,x,'b')
axes[1].set_title('Second Plot')
plt.tight_layout()
```

## Figure size and DPI

```python
fig,axes = plt.subplots(ncols=1,nrows=2,figsize=(8,5))
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('figure.png')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,0.8,0.8])
axes.plot(x,y, label='x vs x squared')
axes.plot(y,x, label='x squared vs x')
axes.set_title('Title')

axes.legend(loc=0)
#axes.legend(loc=(0.1,0.1))
```
